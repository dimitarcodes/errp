```python
correct = epochs[['5','10']].average()
erroneous = epochs[['6','9']].average()
```

```python
errmincorr = mne.combine_evoked([erroneous, correct], weights=[1, -1])
```

```python
fcz = errmincorr.get_data(picks=['FCz'])
plt.plot(fcz.T)
```
